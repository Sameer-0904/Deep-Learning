```python
import time

def fib(n, d):
    """Compute the nth Fibonacci number using memoization.

    Args:
        n (int): The position in the Fibonacci sequence.
        d (dict): A dictionary to store previously computed Fibonacci numbers.

    Returns:
        int: The nth Fibonacci number.
    """
    if n in d:
        return d[n]
    if n <= 1:
        return n
    d[n] = fib(n - 1, d) + fib(n - 2, d)
    return d[n]
```

```python
start = time.time()
d = {0:1, 1:1}
fib(10, d)
end = time.time()
print(end - start)
```
